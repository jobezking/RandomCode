```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("iris.csv")
df2 = pd.read_csv("iris.csv")
```

```python
X = df.iloc[:, :4].values 
X2 = df.iloc[:, [0, 1, 2, 3]].values 
```

```python
print(X)
```

```python
print(X2)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
y_kmeans = kmeans.fit_predict(X)
y_kmeans2 = kmeans.fit_predict(X2)
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(X[y_kmeans == 0, 2], X[y_kmeans == 0, 3], s=100, c='purple', label='Cluster 1')
plt.scatter(X[y_kmeans == 1, 2], X[y_kmeans == 1, 3], s=100, c='blue', label='Cluster 2')
plt.scatter(X[y_kmeans == 2, 2], X[y_kmeans == 2, 3], s=100, c='green', label='Cluster 3')
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(X2[y_kmeans2 == 0, 2], X2[y_kmeans2 == 0, 3], s=100, c='purple', label='Cluster 1')
plt.scatter(X2[y_kmeans2 == 1, 2], X2[y_kmeans2 == 1, 3], s=100, c='blue', label='Cluster 2')
plt.scatter(X2[y_kmeans2 == 2, 2], X2[y_kmeans2 == 2, 3], s=100, c='green', label='Cluster 3')
```
